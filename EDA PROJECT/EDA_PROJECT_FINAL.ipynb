#GRAPHICAL REPRESENTATION



Understanding the data in columns by looking at the different plots.

```python
column_plot = ["Actual_Shipment_Time", "Planned_Shipment_Time", "Planned_Delivery_Time",
                   "Carrier_Num", "Planned_TimeofTravel", "Shipment_Delay", "Distance", "Delivery_Status"]
for column in column_plot:
    plt.figure(figsize=(8, 6))
    plt.hist(data[column], bins=1000, color='yellow', alpha=0.7)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
import seaborn as sns

columns_to_plot = ["Actual_Shipment_Time", "Planned_Shipment_Time", "Planned_Delivery_Time",
                   "Carrier_Num", "Planned_TimeofTravel", "Shipment_Delay", "Distance", "Delivery_Status"]
for column in columns_to_plot:
    plt.figure(figsize=(8, 6))
    sns.kdeplot(data[column], bw = 0.5 , fill = True)
    plt.title(f'Density Plot of {column}')
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.show()
```

```python
columns_plot = ["Actual_Shipment_Time", "Planned_Shipment_Time", "Planned_Delivery_Time",
                   "Carrier_Num", "Planned_TimeofTravel", "Shipment_Delay", "Distance", "Delivery_Status"]
for column in columns_plot:
    plt.figure(figsize=(8, 6))
    sns.boxplot(data[column], color='blue')
    plt.title(f'Box Plot of {column}')
    plt.ylabel(column)
    plt.show()
```
